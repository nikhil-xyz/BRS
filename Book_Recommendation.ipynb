rearranging the titles

```python
def preprocess(title):
  index = title.find(',')
  if (index != -1):
    title = title[index+2:] + " " + title[:index]
  return title
books_new['Title'] = books_new['Title'].apply(preprocess)
```

```python
books_new.Title.head(50)
```

```python
from scipy import sparse
from sklearn.metrics.pairwise import cosine_similarity
```

```python
table = pd.pivot_table(data=books_new, index='Title', columns='SubGenre')
table
```

```python
table = table.fillna(0)
table[table != 0] = 1
table
```

```python
similarity_score = cosine_similarity(table)
similarity_score
```

```python
import numpy as np
def recommend(book_name):
    index = np.where(table.index == book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_score[index])), key=lambda x:x[1], reverse=True)[1:6]

    for i in similar_items:
        print(table.index[i[0]])
```

```python
recommend('Data Mining Handbook')
```
